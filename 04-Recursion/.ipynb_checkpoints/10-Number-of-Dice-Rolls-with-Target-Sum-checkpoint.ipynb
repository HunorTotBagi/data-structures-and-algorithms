```python
def dice(p, target):
    if target == 0:
        print(p)
        return
    
    i = 1
    while i <= 6 and i <= target:
        dice(p+str(i), target-i)
        i += 1
        
dice(" ",5)
```

```python
# Dice with custom number of faces
def diceFace(p, target, face):
    if target == 0:
        print(p)
        return
    
    i = 1
    while i <= face and i <= target:
        diceFace(p+str(i), target-i, face)
        i += 1
        
diceFace(" ",9,7)
```
